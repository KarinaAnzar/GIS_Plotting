```python
import geopandas as gpd
import os
import fiona
import matplotlib.pyplot as plt
```

```python
#importing shape files from geodatabase
gdb_path = r"C:\Users\karis\ArcGISPro\Arcpy_class files\geoportal.gdb (1)\geoportal.gdb"
layers = fiona.listlayers(gdb_path)
```

```python
for layer in layers:
    print(layer)
```

```python
#plotting all the shape files
for layer in layers:
    gdf = gpd.read_file(gdb_path, layer = layer)
    fig, ax = plt.subplots(figsize = (6,6),layout = 'constrained')
    gdf.plot(ax = ax)
    ax.set_title(f"Layer: {layer}")
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    plt.show()
```

```python
## Create a single figure to plot all layers

gdb_path = r"C:\Users\karis\ArcGISPro\Arcpy_class files\geoportal.gdb (1)\geoportal.gdb" 
layers = ['Community_Garden', 'County', 'Farmer_Market', 'Fire', 'Grocer','Health_Centers', 'Hospital', 'Law_Enforcement','Library','Parks','Rail','School'] 
fig, ax = plt.subplots(figsize=(10, 10))

# Create a list to store legend patches
legend_patches = []

# Looping over all layers and plotting them
for layer in layers:  
    gdf = gpd.read_file(gdb_path, layer=layer, layout = 'constrained') # Read the current layer from the geodatabase
    gdf.plot(ax=ax, label=layer, edgecolor='black', alpha=0.5)   # Plot the GeoDataFrame on the same axis

# Setting title and labels
ax.set_title('All Layers in One Figure')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Display a legend 
ax.legend(title='Layers')

# Show the plot
plt.show()
```

```python
#All layers in one frame

gdb_path = r"C:\Users\karis\ArcGISPro\Arcpy_class files\geoportal.gdb (1)\geoportal.gdb"
layers = ['Community_Garden', 'County', 'Farmer_Market', 'Fire', 'Grocer','Health_Centers', 'Hospital', 'Law_Enforcement','Library','Parks','Rail','School'] 

# Create subplots with one row and two columns 
num_layers = len(layers)
columns = 2   
rows = (num_layers//columns)

# Create a figure and axes for subplots
fig, axes = plt.subplots(rows, columns, figsize=(12, 6 * rows), layout = 'constrained')  
axes = axes.flatten()  # Flatten the axes array to make indexing easier

# Loop over all layers and plot them in the corresponding subplot
for i, layer in enumerate(layers):
    gdf = gpd.read_file(gdb_path, layer=layer)  
    
    # Plot the GeoDataFrame in the corresponding subplot
    ax = axes[i]  # Get the appropriate subplot axis
    gdf.plot(ax = ax)
    ax.set_title(f"Layer: {layer}")
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')

plt.tight_layout() # Adjust layout for better spacing

# Show the plot
plt.show()
```
